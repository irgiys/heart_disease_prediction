# Import package yang diperlukan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Load datasets

```python
df = pd.read_csv("health-dataset/hypertension_data.csv")
```

# Exploratory Data Analisis

```python
# Cek / hitung data null pada tiap-tiap kolom
df.isnull().sum()
```

```python
# Hapus data null
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
# Hapus data duplikat
df.drop_duplicates(inplace=True)
```

```python
# Membuat temporary dataframe
temp_df = df.copy()
```

```python
# Ubah data numerik menjadi kategorikal
temp_df["target"] = temp_df["target"].map({0: "Normal", 1:"Heart Disease"})
temp_df["sex"] = temp_df["sex"].map({0:"Female", 1:"Male"})
temp_df['cp']= temp_df['cp'].map({0:'typical angina',1:'atypical angina',2:'non-anginal pain',3:'asymptomatic'})
temp_df['restecg']= temp_df['restecg'].map({0:'Normal',1:'ST-T abnormal',2:'Left ventricular'})
temp_df['exang'] = temp_df['exang'].map({0:'No',1:'Yes'})
temp_df["slope"] = temp_df["slope"].map({0:'upsloping', 1:'flat',2:'downsloping'})
# karena kurangnya penjelasan pada dataset jadi untuk nilai 0 tidak dimasukan
temp_df["thal"] = temp_df["thal"].map({1:"Normal", 2:"Fixed defect", 3:"Reversable"})
```

```python
# Hitung persentase pasien yang memiliki atau tidak sakit jantung
countNoDisease = len(df[df.target == 0])
countHaveDisease = len(df[df.target == 1])
print("Persentase pasien tidak memiliki sakit jantung: {:.2f}%".format((countNoDisease / (len(df.target))*100)))
print("Percentase pasien memiliki sakit jantung: {:.2f}%".format((countHaveDisease / (len(df.target))*100)))
```

# Feature

```python
X = df.drop(columns="target", axis=1)
Y = df["target"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X.shape, X_train.shape
```

# Membuat model

```python
# Disini menggunakan max_iter untuk menghilangkan ConvergenceWarning
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Akurasi data training : ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test )
```

```python
print("Akurasi data test : ", test_data_accuracy)
```

```python
# kena
# input_data =  np.array([57.0,1.0,3,145,233,1,0,150,0,2.3,0,0,1])
# tidak
input_data =  np.array([58.0,0.0,1,120,284,0,0,160,0,1.8,1,0,3])

input_data_reshaped = input_data.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)
if (prediction[0] == 0) :
    print("Pasien tidak terkena penyakit jantung")
else :
    print("Pasien terkena penyakit penyakit jantungi")  
```
